```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab.patches import cv2_imshow
```

```python


original_image = cv2.imread("/content/noisy_image.png",cv2.IMREAD_GRAYSCALE)

```

```python
plt.subplot(121), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
```

```python
# Apply the Averaging Filter
kernel_size = (9, 9)
averaging_filter = np.ones(kernel_size, np.float32) / (kernel_size[0] * kernel_size[1])
averaged_image = cv2.filter2D(original_image, -1, averaging_filter)

# Display the original and averaged images
plt.subplot(121), plt.imshow(original_image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(averaged_image, cmap='gray'), plt.title('Averaged Image')
plt.show()
```

```python
# Apply Gaussian filter
kernel_size = (5, 5)
sigma = 1.5
filtered_image = cv2.GaussianBlur(original_image, kernel_size, sigma)
# Display the original and filtered images side by side
plt.subplot(1, 2, 1), plt.imshow(original_image, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(1, 2, 2), plt.imshow(filtered_image, cmap='gray')
plt.title('Gaussian Filtered Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


original_image = cv2.imread(r"/content/blurred.tif", cv2.IMREAD_GRAYSCALE)

        # High boost filter parameters
k = 25  # Boosting factor     # i suppose its like that,change it as you like <3
kernel_size = (9, 9)  # Kernel size for smoothing

# Apply high boost filter
smoothed_image = cv2.GaussianBlur(original_image, kernel_size, 0)
high_boost_image = cv2.addWeighted(original_image, 1 + k, smoothed_image, -k, 0)

# Apply Canny edge detector on original image
edges_original = cv2.Canny(original_image, 100, 200)

# Apply Canny edge detector on high-boosted image
edges_high_boost = cv2.Canny(high_boost_image, 100, 200)

# Display results
cv2_imshow(original_image)
cv2_imshow(high_boost_image)
cv2_imshow(edges_original)
cv2_imshow(edges_high_boost)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Save the results if needed
cv2.imwrite('edges_original.jpg', edges_original)
cv2.imwrite('edges_high_boost.jpg', edges_high_boost)
```
